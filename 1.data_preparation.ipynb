```python
import os
import shutil

import pandas as pd

from cv2 import cv2
from glob import glob
from tqdm import tqdm, trange
from keras.models import load_model
```

```python
files = glob('data/labeled/*/*.*')
files
```

```python
data = pd.DataFrame()
names = []
paths = []
for item in files:
    _, name, _ = item.split('\\')
    names.append(name)
    paths.append(item.replace('\\', '/'))
data['name'] = names
data['path'] = paths
data
```

```python
data.to_csv('data/labeled/data.csv', index=False)
```

```python
data = pd.read_csv('data/labeled/data.csv')
data
```

```python
#resize to 160x160
resized_data = pd.DataFrame()
names = []
paths = []
for i in trange(len(data)):
    try:
        os.mkdir('data/labeled/resized/{}'.format(data.iloc[i]['name']))
    except:
        shutil.rmtree('data/labeled/resized/{}'.format(data.iloc[i]['name']))
        os.mkdir('data/labeled/resized/{}'.format(data.iloc[i]['name']))
for i in trange(len(data)):
    try:
        image = cv2.imread(data.iloc[i]['path'])
        image = cv2.resize(image, (160,160))
        cv2.imwrite('data/labeled/resized/{}/{}'.format(data.iloc[i]['name'], data.iloc[i]['path'].split('/')[-1]), image)
        names.append(data.iloc[i]['name'])
        paths.append('data/labeled/resized/{}/{}'.format(data.iloc[i]['name'], data.iloc[i]['path'].split('/')[-1]))
    except:
        print(data.iloc[i]['path'])
        break
resized_data['name'] = names
resized_data['path'] = paths
resized_data
```

```python
resized_data.to_csv('data/labeled/data_resized.csv')
```
